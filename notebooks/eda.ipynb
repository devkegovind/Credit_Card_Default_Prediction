**UNIVARIATE ANALYSIS**

```python
plt.figure(figsize=(35,25))
for i in enumerate(df.columns):
    plt.subplot(6, 5, i[0]+1)
    plt.subplots_adjust(hspace=0.5, wspace=0.3)
    plt.hist(df[i[1]], bins = 10, edgecolor = 'black')
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title(i[1])
plt.show()
```

```python
plt.figure(figsize=(35,25))
for i in enumerate(df.columns):
    plt.subplot(6, 5, i[0]+1)
    plt.subplots_adjust(hspace=0.5, wspace=0.3)
    plt.boxplot(df[i[1]])    
    plt.title(i[1])
plt.show()
```

```python
df = df.reset_index(drop=True)
plt.figure(figsize=(35,25))
for i in enumerate(df.columns):
    plt.subplot(6,5,i[0]+1)
    plt.subplots_adjust(hspace=0.5, wspace=0.3)
    sns.kdeplot(df[i[1]])
    plt.title(i)
    #plt.xlabel('Value')
    #plt.ylabel('Frequency')
    #plt.tight_layout()
plt.show()
```

**BIVARIATE ANALYSIS**

```python
df = df.reset_index(drop=True)
plt.figure(figsize=(35,15))
for i in enumerate(df.columns):    
    plt.subplot(6,5,i[0]+1)
    plt.subplots_adjust(hspace=0.5, wspace=.3)
    plt.scatter(y = df[i[1]], x = df['default.payment.next.month'])
    plt.title(i[1])
   

plt.show()
```

**MULTIVARIANTE ANALYSIS**

```python
plt.figure(figsize=(45,25))
sns.heatmap(df.corr(), annot = True, cbar=False)
```

```python
plt.figure(figsize=(35,25))
for i in enumerate(df.columns):
    plt.subplot(6, 5, i[0]+1)
    plt.subplots_adjust(hspace=0.5, wspace=0.3)
    sns.barplot(df[i[1]])
    plt.xlabel('Values')
    plt.ylabel('Frequency')
    plt.title(i[1])
plt.show()
```
